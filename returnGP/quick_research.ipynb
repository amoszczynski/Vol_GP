```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv("spy_minute.csv", index_col=0)
df.index = pd.to_datetime(df.index)
```

```python
days = df.groupby(df.index.date)
```

```python
returns = []
diffs = []

for day in days:
    minus_30 = day[1][day[1].index.time == datetime.time(15,30)]["close"].values[0]
    eod = day[1].iloc[-1]["close"]
    
    returns.append((eod - minus_30) / minus_30)
    diffs.append(eod - minus_30)
```

```python
import matplotlib.pyplot as plt

plt.hist(returns, bins=10)
plt.show()
```
